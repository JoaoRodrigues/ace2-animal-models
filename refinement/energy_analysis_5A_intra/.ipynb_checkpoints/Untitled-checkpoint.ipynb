```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
sns.set(style="white")
```

```python
df = pd.read_csv('agg_hs.csv', index_col='species')
df[df > 10] = 10.0
# df = df.apply(np.log2)
df.head()
```

```python
ace2_cols = [c for c in df.columns if 'B' in c]
ace2 = df[ace2_cols]
vrbd_cols = [c for c in df.columns if 'E' in c]
vrbd = df[vrbd_cols]
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
mask = ace2[ace2 == 0.0].astype(bool)
sns.heatmap(
    ax=ax,
    data=ace2[mask],
    linewidths=.5,
    cmap="RdBu_r",
    xticklabels=1,
    yticklabels=1,
    cbar_kws={'label': 'Residue Interaction Energy (a.u.)'}
)
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
mask = vrbd[vrbd == 0.0].astype(bool)
sns.heatmap(
    ax=ax,
    data=vrbd[mask],
    linewidths=.5,
    cmap="RdBu_r",
    xticklabels=1,
    yticklabels=1,
    cbar_kws={'label': 'Residue Interaction Energy (a.u.)'}
)
```

```python
# Normalize by human structure
nACE2 = ace2 - ace2.mean()
fig, ax = plt.subplots(figsize=(11, 9))
mask = ace2[ace2 == 0.0].astype(bool)
sns.heatmap(
    ax=ax,
    data=nACE2[mask],
    linewidths=.5,
    cmap="RdBu",
    xticklabels=1,
    yticklabels=1,
    cbar_kws={'label': 'Residue Interaction Energy (a.u.)'}
)
```

```python
# Normalize by human structure
nVRBD = vrbd - vrbd.loc['homo_sapiens']
fig, ax = plt.subplots(figsize=(11, 9))
mask = vrbd[vrbd == 0.0].astype(bool)
sns.heatmap(
    ax=ax,
    data=nVRBD[mask],
    linewidths=.5,
    cmap="RdBu",
    xticklabels=1,
    yticklabels=1,
    cbar_kws={'label': 'Residue Interaction Energy (a.u.)'}
)
```
