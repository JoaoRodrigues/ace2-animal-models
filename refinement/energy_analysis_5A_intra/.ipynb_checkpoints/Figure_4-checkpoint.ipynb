```python
def cm2inch(*tupl):
    inch = 2.54
    if isinstance(tupl[0], tuple):
        return tuple(i/inch for i in tupl[0])
    else:
        return tuple(i/inch for i in tupl)
```

```python
# Scientific to Common
namedict1 = {
    'Homo sapiens': 'Human',
    'Anas platyrhynchos': 'Duck',
    'Bos taurus': 'Cow',
    'Camelus dromedarius': 'Dromedary',
    'Canis lupus familiaris': 'Dog',
    'Capra hircus': 'Goat',
    'Carassius auratus': 'Goldfish',
    'Cavia porcellus': 'Guinea pig',
    'Columba livia': 'Pidgeon',
    'Crocodylus porosus': 'Crocodile',
    'Equus asinus': 'Donkey',
    'Equus caballus': 'Horse',
    'Erinaceus europaeus': 'Hedgehog',
    'Felis catus': 'Cat',
    'Gallus gallus': 'Chicken',
    'Macaca mulatta': 'Macaque',
    'Manis javanica': 'Pangolin',
    'Mesocricetus auratus': 'Hamster',
    'Mus musculus': 'Mouse',
    'Mustela putorius furo': 'Ferret',
    'Oryctolagus cuniculus': 'Rabbit',
    'Ovis aries': 'Sheep',
    'Paguma larvata': 'Civet',
    'Pan troglodytes': 'Chimpanzee',
    'Panthera tigris altaica': 'Siberian Tiger',
    'Pongo abelii': 'Orangutan',
    'Rattus norvegicus': 'Rat',
    'Rhinolophus sinicus': 'Horseshoe Bat',
    'Serinus canaria': 'Canary',
    'Sus scrofa': 'Pig'
}

# Lower case and underscore
namedict = {
    '_'.join(k.lower().split()[:2]): v
    for k, v in namedict1.items()
}
```

```python
rootdir = pathlib.Path('.')
```

```python
# Read in PDB files for each species
parser = PDBParser(QUIET=1)
pdbs = list(rootdir.rglob('*best10_*.pdb'))
pdbdict = collections.defaultdict(dict)
for pdb in pdbs:
    if pdb.parent.name not in pdb.name:
        continue
    print(pdb)
    chaindict = {}
    s = parser.get_structure('x', str(pdb))
    for chain in s.get_chains():
        chaindict[chain.id] = {}
        for res in chain:
            resi = res.id[1]
            resn = protein_letters_3to1.get(res.resname.capitalize())
            if resn is None:
                continue
            chaindict[chain.id][resi] = resn
    species = namedict.get(pdb.parent.name)
    if species is None:
        raise Exception(species)
    pdbdict[species].update(chaindict)
```

```python
df = pd.read_csv(rootdir / 'agg_hs.csv', index_col='species')
df[df > 10] = 10.0
df.head(10)
```

```python
# Let's normalize each column so the heatmap coloring help better
def zscore(values):
    average = values.mean()
    sd = values.std()
    return (average - values) / sd
df = df.apply(zscore, result_type='broadcast')
df.head()
```

```python
# Load seq_id to sort by HS
df2 = pd.read_csv('../seq_id.dat', index_col='Species', decimal=',')
df2.index = map(namedict1.get, df2.index)
df2.head()
```

```python
df.index = map(namedict.get, df.index)
```

```python
df['HS'] = df2['HS']
df.sort_values(by='HS', inplace=True)
df.drop(columns=['HS'], inplace=True)
df.head()
```

```python
df.sum(axis=1).sort_values()
```

```python
ace2_cols = [c for c in df.columns if 'B' in c]
ace2 = df[ace2_cols]
vrbd_cols = [c for c in df.columns if 'E' in c]
vrbd = df[vrbd_cols]
```

```python
# Make dataframe with residue names for annotation
ace2_annot = ace2.copy(deep=True)
for row in ace2_annot.index:
    for col in ace2_annot.columns:
        chain = col[0]
        resid = int(col[1:])
        try:
            aakey = pdbdict[row][chain][resid]
        except KeyError:
            aakey = 'X'
        ace2_annot.loc[row, col] = aakey
ace2_annot.head()
```

```python
# Make dataframe with residue names for annotation
# vrbd = vrbd.loc[['Pangolin', 'Ferret', 'Siberian Tiger', 'Horseshoe Bat', 'Civet', 'Cat', 'Human', 'Hamster', 'Pig', 'Chicken', 'Duck', 'Mouse']]
vrbd_annot = vrbd.copy(deep=True)
for row in vrbd_annot.index:
    for col in vrbd_annot.columns:
        chain = col[0]
        resid = int(col[1:])
        try:
            aakey = pdbdict[row][chain][resid]
        except KeyError:
            aakey = 'X'
        vrbd_annot.loc[row, col] = aakey
vrbd_annot.head()
```

## Figure 3

```python
grid_kws = {"height_ratios": (.9, .05), "hspace": .3}
fig, (ax1, ax2) = plt.subplots(
    nrows=2, ncols=1,
    figsize=cm2inch(20, 15),  # w,h
    dpi=300,
    gridspec_kw=grid_kws
)

mask = ace2[ace2 == 0.0].astype(bool)
hm = sns.heatmap(
    ax=ax1,
    data=ace2[mask],
    annot=ace2_annot[mask].values,
    fmt='s',
    linewidths=0.5,
    cmap="RdBu_r",
    xticklabels=1,
    yticklabels=1,
    cbar_ax=ax2,
    cbar_kws={
        'orientation': 'horizontal',
    },
    annot_kws={
        'size': 6
    }
)

cbar = ax1.collections[0].colorbar
ax2.tick_params(labelsize=6)
cbar.set_label('Residue Interaction Energy (a.u.)', size=6)

ax1.xaxis.set_ticks_position('none')
ax1.yaxis.set_ticks_position('none')

xlabels = [
    c[1:] for c in ace2.columns  # clip chain name
]
ax1.set_xticklabels(
    xlabels,
    fontsize=8
)

ax1.set_yticklabels(
    df.index,
    fontsize=8
);

# fig.suptitle('Energetics of ACE2 Interface Residues', fontsize=10);

# fig.savefig('Figure_3.eps', dpi=300);
# fig.savefig('Figure_3.pdf', dpi=300);
```
